```python
def get_time_bins(val,edges):
    if np.isnan(val):
        return np.nan
    time_bin = np.where(val<=edges)[0][0]
    return time_bin
```

```python
age_bins = np.array([1, 10, 20, 30, 40, 50, 60, 70, 80, 90])
df_all['AGE_BIN'] = df_all['ADMIT_AGE'].apply(get_time_bins, args=(age_bins,))
```
